```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

#Data Exploration
df = pd.read_csv('./data/pulsar_stars.csv')
df.head()
```

```python
#Renaming columns:
cust_names = {' Mean of the integrated profile': 'mean_IP', 
             ' Standard deviation of the integrated profile': 'sd_IP',
             ' Excess kurtosis of the integrated profile':'kurt_IP',
             ' Skewness of the integrated profile':'skew_IP',
             ' Mean of the DM-SNR curve':'mean_curve',
             ' Standard deviation of the DM-SNR curve':'sd_curve',
             ' Excess kurtosis of the DM-SNR curve':'kurt_curve',
             ' Skewness of the DM-SNR curve':'skew_curve',
             'target_class':'target'}
```

```python
df.rename(columns=cust_names, inplace=True)
df.columns
```

```python
#Checking the distributions of all features present in the input dataset
fig, axes = plt.subplots(nrows=3,ncols=3,figsize=(10,10))
for ind, column in enumerate(df.columns):
    sns.distplot(df[column],ax=axes[ind//3,ind%3])
    
```

```python
# Checking normality of the feature by plotting qq-line
import pylab
import scipy.stats as stats

fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15,15))
for i, column in enumerate(df.columns):
    res_plt = stats.probplot(df[column],dist='norm',plot=plt)
    fig.add_subplot(3,3,i+1)
    
plt.show()
```
